# Poisson Distribution

Using scipy.stats.poisson to compute PMF and CDF.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

lam = 3
k = np.arange(0, 15)
pmf = poisson.pmf(k, mu=lam)
plt.bar(k, pmf)
plt.xlabel('k')
plt.ylabel('P(X=k)')
plt.title('Poisson PMF (lambda=3)')
plt.show()
```
