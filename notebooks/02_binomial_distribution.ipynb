# Binomial Distribution

Using scipy.stats.binom to compute PMF and CDF.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

n, p = 10, 0.5
k = np.arange(0, n+1)
pmf = binom.pmf(k, n, p)
plt.bar(k, pmf)
plt.xlabel('k')
plt.ylabel('P(X=k)')
plt.title('Binomial PMF (n=10, p=0.5)')
plt.show()
```
